```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
#between male and female who uses home/lifestyle products
super_productline2=df[(df.product_line=="Home and lifestyle")]
super_homeandlifestyle = super_productline2.groupby("Gender")["Quantity"].sum().sort_values(ascending=False)
super_homeandlifestyle
```
